# Exploratory Data Analysis — Real Estate Price Prediction (India Focus)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams["figure.dpi"] = 100
```

## 1. Load Dataset

```python
df = pd.read_csv("../data/train.csv")
print(f"Shape: {df.shape}")
df.head(10)
```

## 2. Dataset Info & Summary Statistics

```python
df.info()
```

```python
df.describe(include="all").T
```

## 3. Missing Values Analysis

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({"Missing Count": missing, "Missing %": missing_pct})
missing_df[missing_df["Missing Count"] > 0].sort_values("Missing %", ascending=False)
```

```python
plt.figure(figsize=(12, 5))
sns.heatmap(df.isnull(), cbar=True, yticklabels=False, cmap="viridis")
plt.title("Missing Values Heatmap", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

## 4. Target Variable Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df["price"].dropna(), kde=True, bins=50, ax=axes[0], color="#667eea")
axes[0].set_title("Price Distribution", fontsize=13, fontweight="bold")
axes[0].set_xlabel("Price (Lakhs)")

sns.boxplot(x=df["price"].dropna(), ax=axes[1], color="#764ba2")
axes[1].set_title("Price Box Plot", fontsize=13, fontweight="bold")
axes[1].set_xlabel("Price (Lakhs)")

plt.tight_layout()
plt.show()
```

## 5. Numerical Feature Distributions

```python
numerical_cols = ["property_area", "bedrooms", "bathrooms", "year_built", "garage_size", "lot_area"]

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
for idx, col in enumerate(numerical_cols):
    ax = axes[idx // 3, idx % 3]
    sns.histplot(df[col].dropna(), kde=True, bins=30, ax=ax, color="#667eea")
    ax.set_title(f"{col} Distribution", fontsize=12, fontweight="bold")

plt.tight_layout()
plt.show()
```

## 6. Categorical Feature Distributions

```python
categorical_cols = ["location", "property_type", "furnishing_status"]

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for idx, col in enumerate(categorical_cols):
    sns.countplot(data=df, x=col, ax=axes[idx], palette="muted", order=df[col].value_counts().index)
    axes[idx].set_title(f"{col} Distribution", fontsize=12, fontweight="bold")
    axes[idx].tick_params(axis="x", rotation=45)

plt.tight_layout()
plt.show()
```

## 7. Correlation Heatmap

```python
plt.figure(figsize=(10, 8))
corr = df[numerical_cols + ["price"]].corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap="coolwarm",
            linewidths=0.5, square=True, vmin=-1, vmax=1)
plt.title("Feature Correlation Heatmap", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

## 8. Price vs Key Features

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.scatterplot(data=df, x="property_area", y="price", alpha=0.4, ax=axes[0], color="#667eea")
axes[0].set_title("Price vs Property Area", fontsize=12, fontweight="bold")

sns.boxplot(data=df, x="bedrooms", y="price", ax=axes[1], palette="muted")
axes[1].set_title("Price vs Bedrooms", fontsize=12, fontweight="bold")

sns.boxplot(data=df, x="location", y="price", ax=axes[2], palette="muted",
            order=["Metro", "Tier1", "Tier2", "Tier3"])
axes[2].set_title("Price vs Location Tier", fontsize=12, fontweight="bold")

plt.tight_layout()
plt.show()
```

## 9. Key Insights

- **Property area and lot area** are the strongest predictors of price
- **Metro locations** command significantly higher prices compared to Tier 2/3 cities
- **Villas** are priced higher than apartments and independent houses
- **Furnished properties** show a clear premium over unfurnished ones
- Missing values are minimal (~3%) and can be handled with median/mode imputation
